```python
import numpy as np
import pandas as pd
import glob
file_path = "C:/Users/bakht/OneDrive - University of Aberdeen/BT5511/Biological computation/Biological computation data/Urna8_data_file.txt" 
df = pd.read_csv(file_path, sep="\t")
folder_path = "C:/Users/bakht/OneDrive - University of Aberdeen/BT5511/Biological computation/Biological computation data/*.txt"
file_list = glob.glob(folder_path)
for file_path in file_list:
    df = pd.read_csv(file_path, sep="\t")
    print(f"{file_path} has {df.shape[0]} rows")
```

```python
df_list = []

for file_path in file_list:
    df = pd.read_csv(file_path, sep="\t")  
    df_list.append(df)

df = pd.concat(df_list, ignore_index=True)
df.index = range(1, len(df) + 1)

print("Number of genes loaded:", len(df))
print(df.head(49))
```

```python
mRNA_WT_Mean = np.mean(df[["mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3"]].values, axis=1)
mRNA_Mut_Mean = np.mean(df[["mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3"]].values, axis=1)

Protein_WT_Mean = np.mean(df[["Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"]].values, axis=1)
Protein_Mut_Mean = np.mean(df[["Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3"]].values, axis=1)

CellViability_WT_Mean = np.mean(df[["CellViability.WT.Rep1", "CellViability.WT.Rep2", "CellViability.WT.Rep3"]].values, axis=1)
CellViability_Mut_Mean = np.mean(df[["CellViability.Mut.Rep1", "CellViability.Mut.Rep2", "CellViability.Mut.Rep3"]].values, axis=1)

df["mRNA_WT_Mean"] = mRNA_WT_Mean
df["mRNA_Mut_Mean"] = mRNA_Mut_Mean
df["Protein_WT_Mean"] = Protein_WT_Mean
df["Protein_Mut_Mean"] = Protein_Mut_Mean
df["CellViability_WT_Mean"] = CellViability_WT_Mean
df["CellViability_Mut_Mean"] = CellViability_Mut_Mean
df.index = range(1, len(df) + 1)

print("\n Mean Expression Levels:")
print(df[["Gene", "mRNA_WT_Mean", "mRNA_Mut_Mean", "Protein_WT_Mean", "Protein_Mut_Mean", "CellViability_WT_Mean", "CellViability_Mut_Mean"]].head(49))
```

```python
from difflib import SequenceMatcher

def compare_sequences(wt_seq, mut_seq):
    sm = SequenceMatcher(None, wt_seq, mut_seq)
    changes = [(tag, i1, i2, j1, j2) for tag, i1, i2, j1, j2 in sm.get_opcodes() if tag != 'equal']
    return changes

def classify_mutation(wt, mut):
    if wt == mut:
        return "No Change"
    elif len(wt) < len(mut):
        return "Insertion"
    elif len(wt) > len(mut):
        return "Deletion"
    else:
        return "Substitution"

mutation_summary = []

for index, row in df.iterrows():
    gene_name = row["Gene"]
    wt_seq = row["WildType.Sequence"]
    mut_seq = row["Mutant.Sequence"]
    
    changes = compare_sequences(wt_seq, mut_seq)
    mutation_class = classify_mutation(wt_seq, mut_seq)

    for change in changes:
        tag, i1, i2, j1, j2 = change
        wt_segment = wt_seq[i1:i2]
        mut_segment = mut_seq[j1:j2]
        
        mutation_summary.append({
            "Gene": gene_name,
            "Mutation_Class": mutation_class,
            "WT_Position": f"{i1}:{i2}",
            "Mut_Position": f"{j1}:{j2}",
            "WT_Segment": wt_segment,
            "Mut_Segment": mut_segment})

mutation_df = pd.DataFrame(mutation_summary)
mutation_df.index = range(1, len(mutation_df) + 1)

df["Mutation_Class"] = df.apply(lambda row: classify_mutation(row["WildType.Sequence"], row["Mutant.Sequence"]), axis=1)
df.index = range(1, len(df) + 1)

print("Mutation summary preview:")
print(mutation_df.head(49))
```

```python
df["mRNA_WT_Mean"] = df[["mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3"]].mean(axis=1)
df["mRNA_Mut_Mean"] = df[["mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3"]].mean(axis=1)

df["Protein_WT_Mean"] = df[["Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"]].mean(axis=1)
df["Protein_Mut_Mean"] = df[["Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3"]].mean(axis=1)

mRNA_ratio = df["mRNA_Mut_Mean"] / df["mRNA_WT_Mean"]
mRNA_ratio = mRNA_ratio.replace([np.inf, -np.inf], np.nan)
mRNA_ratio[mRNA_ratio <= 0] = np.nan
df["mRNA_Log2FC"] = np.log2(mRNA_ratio)

protein_ratio = df["Protein_Mut_Mean"] / df["Protein_WT_Mean"]
protein_ratio = protein_ratio.replace([np.inf, -np.inf], np.nan)
protein_ratio[protein_ratio <= 0] = np.nan
df["Protein_Log2FC"] = np.log2(protein_ratio)

print(f"Number of genes analyzed: {len(df)}\n")

print("Top 5 genes with highest mRNA Log2 Fold Change:")
print(df[["Gene", "mRNA_Log2FC"]].sort_values(by="mRNA_Log2FC", ascending=False).head(5), "\n")

print("Top 5 genes with highest Protein Log2 Fold Change:")
print(df[["Gene", "Protein_Log2FC"]].sort_values(by="Protein_Log2FC", ascending=False).head(5))
```

```python
from scipy.stats import ttest_rel

df["mRNA_WT_Mean"] = df[["mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3"]].mean(axis=1)
df["mRNA_Mut_Mean"] = df[["mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3"]].mean(axis=1)

df["Protein_WT_Mean"] = df[["Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"]].mean(axis=1)
df["Protein_Mut_Mean"] = df[["Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3"]].mean(axis=1)

mRNA_ratio = df["mRNA_Mut_Mean"] / df["mRNA_WT_Mean"]
mRNA_ratio = mRNA_ratio.replace([np.inf, -np.inf], np.nan)
mRNA_ratio[mRNA_ratio <= 0] = np.nan
df["mRNA_Log2FC"] = np.log2(mRNA_ratio)

protein_ratio = df["Protein_Mut_Mean"] / df["Protein_WT_Mean"]
protein_ratio = protein_ratio.replace([np.inf, -np.inf], np.nan)
protein_ratio[protein_ratio <= 0] = np.nan
df["Protein_Log2FC"] = np.log2(protein_ratio)

mRNA_pvals = []
protein_pvals = []

for _, row in df.iterrows():
    wt_mrna = [row["mRNA.Expression.WT.Rep1"], row["mRNA.Expression.WT.Rep2"], row["mRNA.Expression.WT.Rep3"]]
    mut_mrna = [row["mRNA.Expression.Mut.Rep1"], row["mRNA.Expression.Mut.Rep2"], row["mRNA.Expression.Mut.Rep3"]]
    _, pval_mrna = ttest_rel(wt_mrna, mut_mrna, nan_policy='omit')
    mRNA_pvals.append(pval_mrna)

    wt_prot = [row["Protein.Expression.WT.Rep1"], row["Protein.Expression.WT.Rep2"], row["Protein.Expression.WT.Rep3"]]
    mut_prot = [row["Protein.Expression.Mut.Rep1"], row["Protein.Expression.Mut.Rep2"], row["Protein.Expression.Mut.Rep3"]]
    _, pval_prot = ttest_rel(wt_prot, mut_prot, nan_policy='omit')
    protein_pvals.append(pval_prot)

df["mRNA_pval"] = mRNA_pvals
df["Protein_pval"] = protein_pvals

df["mRNA_significant"] = df["mRNA_pval"] < 0.05
df["Protein_significant"] = df["Protein_pval"] < 0.05

print(f"Number of genes analyzed: {len(df)}\n")

print("Top 5 significant mRNA expression changes (p < 0.05):")
print(df[df["mRNA_significant"]][["Gene", "mRNA_Log2FC", "mRNA_pval"]].sort_values(by="mRNA_Log2FC", ascending=False).head(5), "\n")

print("Top 5 significant protein expression changes (p < 0.05):")
print(df[df["Protein_significant"]][["Gene", "Protein_Log2FC", "Protein_pval"]].sort_values(by="Protein_Log2FC", ascending=False).head(5), "\n")

print(f"Number of significant mRNA genes: {df['mRNA_significant'].sum()}")
print(f"Number of significant protein genes: {df['Protein_significant'].sum()}")
```

```python
mutation_txt_files = glob.glob("C:/Users/bakht/OneDrive - University of Aberdeen/BT5511/Biological computation/Biological computation data/*.txt")
if not mutation_txt_files:
    raise FileNotFoundError("Mutation .txt-files are not found!")

mutations = pd.concat([pd.read_csv(file, sep='\t') for file in mutation_txt_files], ignore_index=True)

print("Columns:", mutations.columns)

required_columns = ['chrom', 'pos', 'ref', 'alt', 'Gene', 'WildType.Sequence', 'Mutant.Sequence']
for col in required_columns:
    if col not in mutations.columns:
        mutations[col] = None 

def is_promoter_region(wt_seq, mut_pos, promoter_distance=1000):
    if mut_pos is None:
        return False  
    if mut_pos < promoter_distance:
        return True
    return False

mutation_summary = []

for index, row in mutations.iterrows():
    gene_name = row["Gene"]
    wt_seq = row["WildType.Sequence"]
    mut_pos = row["pos"]  

    if is_promoter_region(wt_seq, mut_pos):
        mutation_summary.append({
            "Gene": gene_name,
            "Mutation_Position": mut_pos,
            "Region": "Promoter Region"})
    else:
        mutation_summary.append({
            "Gene": gene_name,
            "Mutation_Position": mut_pos,
            "Region": "Not in Promoter Region"})

mutation_df = pd.DataFrame(mutation_summary)
mutation_df.index = range(1, len(mutation_df) + 1)

output_file = "C:/Users/bakht/OneDrive - University of Aberdeen/BT5511/Biological computation/Biological computation data/mutation_in_promoter_region.csv"
mutation_df.to_csv(output_file, index=False)
print(f" The results of mutations in the promoter region are preserved: {output_file}")

print("Mutation in Promoter Region preview:")
print(mutation_df.head(49))
```
